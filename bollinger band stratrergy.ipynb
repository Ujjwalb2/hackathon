```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import math
from termcolor import colored as cl
import numpy as np
```

```python
df=pd.read_csv('/content/HDFCBANK.NS.csv')
```

```python
def sma(data, window):
    sma = data.rolling(window = window).mean()
    return sma
df['sma_20'] = sma(df['Close'], 20)
# df.tail()
```

```python
df
```

```python
def bb(data, sma, window):
    std = data.rolling(window = window).std()
    upper_bb = sma + std * 2
    lower_bb = sma - std * 2
    return upper_bb, lower_bb

df['upper_bb'], df['lower_bb'] = bb(df['Close'], df['sma_20'], 20)
df.tail()
```

```python
df['Close'].plot(label = 'CLOSE PRICES', color = 'skyblue')
df['upper_bb'].plot(label = 'UPPER BB 20', linestyle = '--', linewidth = 1, color = 'black')
df['sma_20'].plot(label = 'MIDDLE BB 20', linestyle = '--', linewidth = 1.2, color = 'grey')
df['lower_bb'].plot(label = 'LOWER BB 20', linestyle = '--', linewidth = 1, color = 'black')
plt.legend(loc = 'upper left')
plt.title('hdfc BOLLINGER BANDS')
plt.show()
```

```python

def implement_bb_strategy(data, lower_bb, upper_bb):
    buy_price = []
    sell_price = []
    bb_signal = []
    signal = 0

    for i in range(1,len(data)):
        if data[i-1] > lower_bb[i-1] and data[i] < lower_bb[i]:
            if signal != 1:
                buy_price.append(data[i])
                sell_price.append(np.nan)
                signal = 1
                bb_signal.append(signal)
            else:
                buy_price.append(np.nan)
                sell_price.append(np.nan)
                bb_signal.append(0)
        elif data[i-1] < upper_bb[i-1] and data[i] > upper_bb[i]:
            if signal != -1:
                buy_price.append(np.nan)
                sell_price.append(data[i])
                signal = -1
                bb_signal.append(signal)
            else:
                buy_price.append(np.nan)
                sell_price.append(np.nan)
                bb_signal.append(0)
        else:
            buy_price.append(np.nan)
            sell_price.append(np.nan)
            bb_signal.append(0)

    return buy_price, sell_price, bb_signal

buy_price, sell_price, bb_signal = implement_bb_strategy(df['Close'], df['lower_bb'], df['upper_bb'])
```

```python
import matplotlib.pyplot as plt

# Assuming buy_price and sell_price are defined correctly

# Aligning the arrays using the DataFrame index
df_aligned, buy_price_aligned = df.align(pd.Series(buy_price), axis=0)
df_aligned, sell_price_aligned = df.align(pd.Series(sell_price), axis=0)

df_aligned['Close'].plot(label='CLOSE PRICES', alpha=0.3)
df_aligned['upper_bb'].plot(label='UPPER BB', linestyle='--', linewidth=1, color='black')
df_aligned['sma_20'].plot(label='MIDDLE BB', linestyle='--', linewidth=1.2, color='grey')
df_aligned['lower_bb'].plot(label='LOWER BB', linestyle='--', linewidth=1, color='black')

plt.scatter(df_aligned.index, buy_price_aligned, marker='^', color='green', label='BUY', s=20)
plt.scatter(df_aligned.index, sell_price_aligned, marker='v', color='red', label='SELL', s=20)

plt.title('hdfc BB STRATEGY TRADING SIGNALS')
plt.legend(loc='upper left')
plt.show()
```
